```python
# Importing and Printing Stopwords
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
# Removing Stopwords from Text file
stop_words = set(stopwords.words('english'))

word_tokens = nltk.word_tokenize(corpus)
filtered_sentence = [w for w in word_tokens if not w in stop_words]
filtered_sentence = []
for w in word_tokens:
  if w not in stop_words:
    filtered_sentence.append(w)
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('omw-1.4')
Lemmatizer=WordNetLemmatizer()
steemer=PorterStemmer()
nltk.download('wordnet')
Lemmatizer_words = []
for w in filtered_sentence:
  rootWord = Lemmatizer.lemmatize(w,pos=wordnet.VERB)
  Lemmatizer_words.append(rootWord)
print(Lemmatizer_words)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_words = []
for w in filtered_sentence:
  rootWord = ps.stem(w)
  stemmed_words.append(rootWord)
print(stemmed_words)
```

```python
frequency_dist = nltk.FreqDist(stemmed_words)
frequency_dist
```

```python
large_words=dict([k,v]for k,v in frequency_dist.items() if len(k)>3)
large_words
```

```python
frequency_dist = nltk.FreqDist(large_words)
frequency_dist
frequency_dist.plot(30,cumulative=False)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud(max_font_size=50, max_words=100,background_color="yellow").generate_from_frequencies(frequency_dist)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
```

```python
import pandas as pd
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
values = tfidf_vectorizer.fit_transform(stemmed_words)
```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()
pd.DataFrame(values.toarray(),columns=feature_names)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create a dictionary with words and their corresponding TF-IDF scores
word_scores = {}
for word, score in zip(feature_names, values.toarray()[0]):
    word_scores[word] = score

```

```python
# Generate word cloud using TF-IDF scores as weights
wordcloud = WordCloud(max_font_size=50, max_words=100,width=800, height=400, background_color='green').generate_from_frequencies(word_scores)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
