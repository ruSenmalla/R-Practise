```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Create a dictionary with words and their corresponding TF-IDF scores
word_scores = {}
for word, score in zip(feature_names, values.toarray()[0]):
    word_scores[word] = score
```

```python
# Generate word cloud using TF-IDF scores as weights
wordcloud = WordCloud(max_font_size=50, max_words=100,width=800, height=400, background_color='green').generate_from_frequencies(word_scores)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
corpus2 = open('/content/drive/MyDrive/Colab Notebooks/The Diamond Sutra /Chapter3.txt').read()
```

```python
#Changing all the words in the text file into lower case
corpus2 = corpus2.lower()
```

```python
# Split the document into lines
lines = corpus2.splitlines()
```

```python
# Initialize the PorterStemmer
stemmer = PorterStemmer()

```

```python
# Tokenize the line into words
stemmed_lines = []
for line in lines:
  tokens = nltk.word_tokenize(line)
  stemmed_line = [stemmer.stem(token) for token in tokens]
  stemmed_lines.append(''.join(stemmed_line))
```

```python
# Join the stemmed lines back into a single string
stemmed_document = '\n'.join(stemmed_lines)
```

```python
from nltk.tokenize import word_tokenize
# Tokenize the stemmed document into words
stemmed_words = word_tokenize(stemmed_document)

# Join the stemmed words into a single string
stemmed_words_str = ' '.join(stemmed_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
values = tfidf_vectorizer.fit_transform(stemmed_words)
feature_names = tfidf_vectorizer.get_feature_names_out()
pd.DataFrame(values.toarray(),columns=feature_names)
```

```python
# Create a dictionary with words and their corresponding TF-IDF scores
word_scores = {}
for word, score in zip(feature_names, values.toarray()[0]):
    word_scores[word] = score

# Generate word cloud using TF-IDF scores as weights
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_scores)
```

```python
# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
